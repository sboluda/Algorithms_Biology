## Create nodes

We will now import a **Node** class and and **newick2nodes()** function that converts a *Newick tree* (a file or a string) to a *dictionary of Node objects*.

```python
from tree_nodes import Node, newick2nodes
```

The class *Node* has name, left, right and parent attributes

```python
!head tree_nodes.py -n 12
```

And a *newick2nodes()* function that creates the nodes object

```python
nodes = newick2nodes("((A,B),((C,(D,E)),F));")
nodes
```

Notice that the variable *nodes* is a dictionary that contains the *node number* as key and the *Node objects* as values.

The following gives the node object with id 4

```python
nodes[8]
```

```python
type(nodes[8])
```

You can use the *dir()* funciton to list the attributes of the node object: **name**, **parent**, **left**, and **right**

```python
dir(nodes[8])
```

Notice that in our tree only the leaves have a name

```python
nodes[8].name
```

Intermediate nodes have the empty string as name

```python
nodes[7].name 
```

Who is the parent of node 8?

```python
nodes[8].parent
```

To get the name of the left child of node 7 we can do this:

```python
nodes[nodes[7].left].name
```

Who is the sister leaf of D?

```python
left_child = nodes[7].left
right_child = nodes[7].right
print(nodes[left_child].name, nodes[right_child].name)
```
