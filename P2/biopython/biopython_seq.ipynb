## The Seq object

```python
dna = Seq('AGTCTTATTTATTAG')
dna
```

```python
print(dna)
```

```python
type(dna)
```

Works similar to a *str*

```python
dna[1:4]
```

```python
dna.count('TT')
```

```python
dna.find('AT')
```

```python
dna.lower()
```

But also gets added functionality. For instance there is a proper transcribe() funtion(method)

```python
rna = dna.transcribe()
print(rna)
```

Or we can translate to protein

```python
prot = rna.translate()
print(prot)
```
